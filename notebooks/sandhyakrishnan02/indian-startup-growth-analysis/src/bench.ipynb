# Visualisations

```python
df = pd.read_csv('../input/indian-startup-recognized-by-dpiit/Startup_Counts_Across_India.csv')
df.head().style.background_gradient(cmap='coolwarm')
```

```python
df = df
intended_df_size_in_MB = 256
factor = intended_df_size_in_MB*(2**20)/df.memory_usage(index=True).sum()
if int(factor) > 0:
    df = pd.concat([df]*int(factor), ignore_index=True)
else:
    rowCount = int(df.shape[0]*factor)
    df = df[0:rowCount]
df = df
df.info()
```
