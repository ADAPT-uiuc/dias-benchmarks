```python
# importing the necessary libraries for EDA on Airline Dataset | Kaggle Dataset

import pandas as pd 
import numpy as np
```

```python
# loading the dataset and showing first 5 rows of the dataset

df = pd.read_csv("../input/Airline Dataset.csv")
df.head(5)
```

```python
# Getting the info on the dataset
df.info()
```

```python
#  Top 20 Country with count of Visitors 

df['Country Name'].value_counts().head(20)
```

```python
# Vizualizing the previous cell 

top_20_countries_df =  df['Country Name'].value_counts().head(20)
top_20_countries_df = pd.DataFrame({'Country Name': top_20_countries_df.index, 'Count': top_20_countries_df.values})
top_20_countries_df = top_20_countries_df.sort_values(by='Count', ascending=True)
```

```python
# Ratio between Male Travellers and Female Travellers

Travl_Ratio =  df['Gender'].value_counts()
Travl_Ratio = pd.DataFrame({'Gender': Travl_Ratio.index, 'Count': Travl_Ratio.values})
```

```python
# Total number of Pilots per country 

Pilot_df = df[['Country Name', 'Pilot Name']]
pilot_pivot = pd.pivot_table(Pilot_df, index=['Country Name'], aggfunc='count')
pilot_pivot.rename(columns={'Pilot Name':'Pilot Count'}, inplace=True)
pilot_pivot = pilot_pivot.reset_index()
pilot_pivot.sort_values(by='Pilot Count', ascending=False, inplace=True)
pilot_pivot.head(10)
```

```python
df.head(5)
```

```python
# Shoes the Count of travellers per Flight Status by Country for the entire duration in the dataset

Flight_status_df = df[['Country Name','Flight Status']]
Flight_status_df['FTE'] = 1

# Creating a Pivot table by Country with Count of Travellers per Flight Status  
Flight_status_pivot = pd.pivot_table(Flight_status_df, index=['Country Name'], columns=['Flight Status'], values = ['FTE'], aggfunc='count')
Flight_status_pivot = Flight_status_pivot.reset_index()
Flight_status_pivot.head(11)
```

```python
# Removing the Spaces on the column names and replacing it with (-) 
# Refrencing the dataframe to another variable (df2)

df.columns = df.columns.str.replace(' ', '_')
df2 = df
df2

# Quering on the data based on Airport Country Code (United States) 
US_data = df2.query('Airport_Country_Code == "US"')


US_data
```

```python
#  The Departure Date is an Object, converting it into Date-time format and extracting Quarter, month and Week 

US_data['Date'] = pd.to_datetime(US_data['Departure_Date']) 

# Extract quarter, month, week

US_data['Quarter'] = US_data['Date'].dt.quarter
US_data['Month'] = US_data['Date'].dt.month
US_data['Week'] = US_data['Date'].dt.isocalendar().week
US_data

#  So Now we have a data_set that only has values for United States for further EDA
```

```python
#  Creating a Customer Column based on Departure and Arrival Airport 
US_data['Route'] = US_data['Airport_Name'].str.cat(US_data['Arrival_Airport'], sep=' -to- ')

US_Flight_Route = US_data['Route'].value_counts()
US_Flight_Route = pd.DataFrame({'Route': US_Flight_Route.index, 'Count': US_Flight_Route.values})
Top_20_US_Flight_Route = US_Flight_Route.head(20)
```

```python
Top_20_US_Flight_Route
```

```python
# Figuring out the total number of flight per month 


filtered_df = df[['Airport_Name','Departure_Date','Arrival_Airport','Pilot_Name']]
filtered_df1 = pd.DataFrame(filtered_df)
filtered_df1_no_duplicates = filtered_df1.drop_duplicates()
filtered_df1_no_duplicates['Date'] = pd.to_datetime(filtered_df1_no_duplicates['Departure_Date']) 

# Extracting month
filtered_df1_no_duplicates['Month'] = filtered_df1_no_duplicates['Date'].dt.month
filtered_df1_no_duplicates
```

```python
# Table for Total number of flight per month (This included all flight status)

filtered_df1_no_duplicates
Total_flights = filtered_df1_no_duplicates['Month'].value_counts()
Total_flights = Total_flights.reset_index()
Total_flights.sort_values(by='index', ascending=True, inplace=True)
Total_flights
```

```python
#  Getting the Average number of flights per month 

Average_Flight = np.mean(Total_flights['Month'])
Average_Flight
```
